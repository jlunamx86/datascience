```python
import numpy as np
from numpy.random import binomial
from scipy.stats import binom
from math import factorial
import matplotlib.pyplot as plt
```

```python
def my_binomial(x, n, p):
    return factorial(n)/(factorial(x)*factorial(n-x))*pow(p,x)*pow(1-p,n-x)
```

```python
#validation: 2 heads, 3 toss, 0.5 probability
my_binomial(2,3,0.5)
```

```python
dist = binom(3, 0.5)
#probability mass function
dist.pmf(2)
```

```python
#cummulative density function
#2 or less heads
dist.cdf(2)
```

<h3>Random Generators</h3>

```python
p=0.5
n=3
binomial(n,p)
```

```python
expermients = []
for _ in range(100):
    expermients.append(binomial(n,p))
```

```python
def plot_hist(num_trials):
    values = [0,1,2,3]
    arr = []
    for _ in range(num_trials):
        arr.append(binomial(n,p))
    simulation = np.unique(arr, return_counts=True)[1]/len(arr)
    theorical_dist = [binom(3, 0.5).pmf(k) for k in values]
    plt.bar(values, simulation, color='red')
    plt.bar(values, theorical_dist, alpha=0.5, color='blue')
    plt.title('{} experiments'.format(num_trials))
    plt.show()
```

```python
plot_hist(20)
```

```python
plot_hist(200)
```

<h3>The more trials the closer the theorical and simulation distributions</h3>

```python
plot_hist(200000)
```
