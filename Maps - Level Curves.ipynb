<h2>Dependencies</h2>

```python
from matplotlib import cm #colours
import numpy as np
import matplotlib.pyplot as plt
```

<h2>3D</h2>
<h4>Plot x & y variables in 3-Dimenision. In Red the highest points and in blue the lowest ones</h4>

```python
def f(x,y):
    return np.sin(x) + 2*np.cos(y)

res = 100 #resolution

x = np.linspace(-4,4,num=res)
y = np.linspace(-4,4,num=res)

x, y = np.meshgrid(x,y)

z = f(x,y)

fig, ax = plt.subplots(subplot_kw={"projection":"3d"})

surf = ax.plot_surface(x,y,z, cmap=cm.coolwarm)
fig.colorbar(surf)
```

<h2>Level Curves</h2>
<h4>Sight from top or 2-Dimensions</h4>

```python
fig2, ax2 = plt.subplots()
level_map = np.linspace(np.min(z), np.max(z), num=res)

cp = ax2.contour(x,y,z, levels=level_map, cmpa=cm.coolwarm)
fig2.colorbar(cp)
```

```python
fig2, ax2 = plt.subplots()
level_map = np.linspace(np.min(z), np.max(z), num=res)

cp = ax2.contourf(x,y,z, levels=level_map, cmpa=cm.coolwarm)
fig2.colorbar(cp)
```
