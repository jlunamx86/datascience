<h1>Seaborn</h1>

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['A','B','C'], y=[1,3,2])

plt.show()
```

```python
sns.set(style='darkgrid',palette='muted',font="Verdana",font_scale=1)
sns.barplot(x=['A','B','C'], y=[1,3,2])
plt.show()
```

<h2>Parameters</h2>

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.displot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',y='tip',hue='sex')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',hue='sex')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='kde',legend=True,palette='dark',alpha=0.5)
plt.show()
```

<h2>Distributions</h2>

```python
sns.histplot(data=tips, x='tip',bins=15,cumulative=True)
plt.show()
```

```python
sns.histplot(data=tips, x='tip',bins=15,cumulative=False,hue='sex',stat='probability',multiple='dodge')
plt.show()
```

```python
sns.kdeplot(data=tips,x='tip',hue='sex',shade=True,bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count')
plt.show()
```

```python
sns.displot(data=tips,x='tip',hue='sex',kind='kde',multiple='stack')
plt.show()
```

<h2>Categorical</h2>

```python
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,color="0",marker='p')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,palette='pastel',split=True)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,color="0",marker='p')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,kind='box',col='time')
plt.show()
```

<h2>Relation</h2>

```python
plt.figure(figsize=(8,8))
markers={"Lunch":"D","Dinner":"s"}
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
sns.lineplot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',col='time')
plt.show()
```

<h2>Jointplot - Pairplot</h2>

```python
sns.jointplot(data=tips,x='total_bill',y='tip',hue='sex',kind='hist',marginal_ticks=True,
              marginal_kws=dict(bins=25,fill=False,multiple='dodge'))
plt.show()
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',hue='sex',kind='kde')
plt.show()
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',hue='sex')
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex',palette='dark',corner=True)
plt.show()
```

<h2>Heatmap</h2>

```python
tips.corr()
```

```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm',linewidths=5,linecolor='black',
           vmin=0.5,vmax=1,cbar=True)
plt.show()
```
