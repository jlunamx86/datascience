<h1>Pyplot<h1>

<h2>Basic</h2>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
plt.plot(x,y,'rs-')
plt.show()
```

```python
plt.plot(x,y,'yD:')
plt.show()
```

```python
plt.hist(x)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```

<h2>Subplot</h2>

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.subplot(1, 2, 1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')
plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')
plt.subplot(2,1,2)
plt.pie(y)
plt.show()
```

<h2>Object Oriented</h2>

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
#object
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])
axes.plot(x,y,'b')
axes2.plot(y,x,'r')
axes2.set_facecolor('gray')
fig.show()
```

<h3>Canvas</h3>

<h2>Subplots<h3>

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

```python
fig, axes = plt.subplots(2,4)
axes[0,0].plot(x,np.cos(x))
axes[0,1].plot(x,np.sin(x),'y')
axes[0,2].plot(x,np.tan(x),'r')
axes[0,3].plot(x,np.cos(x)**2)
```

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2,4)
ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x),'y')
ax3.plot(x,np.tan(x),'r')
ax4.plot(x,np.cos(x)**2)
#adjust layout
fig.tight_layout()
```

<h2>Legends, labels, titles, size</h2>

```python
fig, axes = plt.subplots(1,2,figsize=(10,5))
axes[0].plot(x,y,label="$sin(x)$")
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()
axes[1].plot(y,x)
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
fig.tight_layout()
```

```python
plt.plot(x,y,label="$sin(x)$")
plt.title("My Chart")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc="lower right",bbox_to_anchor=(1,0.2))
```

<h2>Colours & Styles</h2>

```python
print(plt.style.available)
```

```python
plt.style.use('ggplot')
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'y.')
ax.plot(x,x+4,'go:')
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,color='green',alpha=0.8,linewidth=5,linestyle='-',marker='o',markersize=10,markerfacecolor='#CCFFCC')
ax.plot(x,x+2,color='blue',linewidth=8,linestyle='--',marker='x')
ax.plot(x,x+3,color='#66FF66',linestyle='dashed',marker='v')
ax.plot(x,x+4,color='yellow',marker='P')
```

<h2>Bar Plot</h2>

```python
country = ['INDIA','JAPAN','MEXICO','COLOMBIA','GERMANY']
population = [1000,800,900,1000,300]
```

```python
plt.bar(country,population,width=0.5,color=['red','blue'])
plt.xticks(np.arange(5),('India','Japan','Mexico','Colombia','Deutschland'),rotation=45)
plt.show()
```

```python
plt.barh(country,population)
plt.show()
```

<h2>Other charts</h2>

```python
data= np.random.randint(1,50,100)
```

```python
plt.hist(data,bins=10,histtype='bar')
plt.show()
```

```python
plt.hist(data,bins=10,histtype='step')

plt.show()
```

```python
#outlier
data = np.append(data,300)
plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=True)
plt.show()
```

```python
plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=False)
plt.show()
```
