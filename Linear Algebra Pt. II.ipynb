<h2>Matrix's properties</h2>

```python
A = np.array([[2,3],[5,7],[11,13]])
B = np.array([[1,3],[2,1]])
C = np.array([[3,1],[4,2]])
```

<h3>Associative</h3>
<p>Ax(BxC) = (AxB)xC</p>

```python
ABC = A.dot(B.dot(C))
ABC
```

```python
AB_C = A.dot(B).dot(C)
AB_C
```

<h3>Distributive</h3>
<p>Ax(B+C) = (AxB)+(AxC)</p>

```python
D = A.dot(B+C)
D
```

```python
E = (A.dot(B)) + (A.dot(C))
E
```

```python
D==E
```

<h3>Commutative</h3>
<p>BxC = CxB</p>

```python
F = B.dot(C)
F
```

```python
G = C.dot(B)
G
```

<h3>Not commutative</h3>

```python
F == G
```

<h3>Vectors are commutative</h3>

```python
v1 = np.array([[2],[7]])
v2 = np.array([[3],[5]])
```

```python
v1_T_v2 = v1.T.dot(v2)
v1_T_v2
```

```python
v2_T_v1 = v2.T.dot(v1)
v2_T_v1
```

```python
v1_T_v2 == v2_T_v1
```

<h2>Tranposition of a matrices' product</h2>
<p>(AxB)trans = (B)trans x (A)trans</p>
<p>((AxB)trans)trans = AxB</p>

```python
A = np.array([[2,3],[5,7],[11,13]])
B = np.array([[1,3],[2,1]])
```

```python
AB_t = A.dot(B).T
AB_t
```

```python
B_TA_T = B.T.dot(A.T)
B_TA_T
```

```python
AB_t == B_TA_T
```

<h2>Use inverse to solve an equation system</h2>
<h3>3X+Y = 1</h3>
<h3>2X+Y = 1</h3>

```python
np.set_printoptions(suppress=True)
```

```python
A = np.array([[3,1],[2,1]])
A
```

```python
B = np.array([[1],[1]])
B
```

```python
inverse_A = np.linalg.inv(A)
inverse_A
```

```python
X = inverse_A.dot(B)
X
```

```python
A.dot(X)
```

<h3>3X+Y=3</h3>
<h3>2X+Y=7</h3>

```python
sol_2 = inverse_A.dot(np.array([[3],[7]]))
sol_2
```

```python
A.dot(sol_2)
```
