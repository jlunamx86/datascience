<h1>Function Manipulation</h1>

Let be C a constant greater than 0:
\begin{gather*}
y = f(x) + C\\
\end{gather*}
It moves C units upwards

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 1000
def f(x):
    return x**2

c = 5
x = np.linspace(-5,5, num=N)
y = f(x) + c

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```

Let be C a constant greater than 0:
\begin{gather*}
y = f(x) - C\\
\end{gather*}
It moves C units downwards

```python
N = 1000
def f(x):
    return x**2

c = 5
x = np.linspace(-5,5, num=N)
y = f(x) - c

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```

Let be C a constant greater than 0:
\begin{gather*}
y = f(x - C) \\
\end{gather*}
It moves c units towards right

```python
N = 1000
def f(x):
    return x**2

c = 5
x = np.linspace(-10,10, num=N)
y = f(x - c)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```

Let be C a constant greater than 0:
\begin{gather*}
y = f(x + C) \\
\end{gather*}
It moves c units towards left

```python
N = 1000
def f(x):
    return x**2

c = 5
x = np.linspace(-10,10, num=N)
y = f(x + c)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```

Let be C a constant greater than 0:
\begin{gather*}
y = C.f(x) \\
\end{gather*}
It vertically enlarges the chart by a c factor

```python
N = 1000
def f(x):
    return np.sin(x)

c = 3
x = np.linspace(-15,15, num=N)
y = c * f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```

Let be C a constant greater than 0:
\begin{gather*}
y = \frac{1}{C} .f(x) \\
\end{gather*}
It vertically compresses the chart by a c factor

```python
N = 1000
def f(x):
    return np.sin(x)

c = 3
x = np.linspace(-15,15, num=N)
y = (1/c) * f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```

Let be C a constant greater than 0:
\begin{gather*}
y = f(C.x)\\
\end{gather*}
It horizontally compresses the chart by a c factor

```python
N = 1000
def f(x):
    return np.sin(x)

c = 2
x = np.linspace(-15,15, num=N)
y = f(c * x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```

Let be C a constant greater than 0:
\begin{gather*}
y = f(\frac{1}{C}.x) \\
\end{gather*}
It horizontally enlarges the chart by a c factor

```python
N = 1000
def f(x):
    return np.sin(x)

c = 2
x = np.linspace(-15,15, num=N)
y = f((1/c) * x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```

\begin{gather*}
y = -f(x)\\
\end{gather*}
It reflects the chart with respect to X axis.

```python
N = 1000
def f(x):
    return x**3

x = np.linspace(-10,10, num=N)
y = -f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```


\begin{gather*}
y = f(-x)\\
\end{gather*}
It reflects the chart with respect to Y axis.

```python
N = 1000
def f(x):
    return x**3

x = np.linspace(-10,10, num=N)
y = f(-x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```
