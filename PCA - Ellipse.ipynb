<h1>PCA (1) applied to an Ellipse</h1>

<h2>Get Component 1 from an ellipse<br>*compare to PCA - circle</h2>

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

<h2>Ellipse equation</h2>

```python
#center=(0,0)
#a = 1
#b = 0.4
def x_ellipse_center_origin(y:float):
    return  math.sqrt(1 - (y**2/0.4**2))
```

<h3>500 random points from a uniform distribution</h3>

```python
lt_coordinates = list()
for _ in range(0,500):
    y = np.random.uniform(-0.4,0.4)    
    lt_coordinates.append((x_ellipse_center_origin(y),y))   
    lt_coordinates.append((-1 * x_ellipse_center_origin(y),y))
```

```python
np_arr = np.array(lt_coordinates)
x, y = np_arr.T
```

```python
plt.scatter(x, y)
plt.show()
```

```python
df_data = pd.DataFrame(np_arr)
df_data.rename(columns={0:"X",1:"Y"},inplace=True)
pca_pipeline = make_pipeline(PCA(n_components=1))
pca_data = pca_pipeline.fit_transform(df_data)
```

<h3>With component one, 92.32% of variance</h3>

```python
sum(pca_pipeline['pca'].explained_variance_ratio_)
```
