<h1>Logistic Regression & Maximum Likelihood Estimation</h1>

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
import pandas as pd
```

```python
def likelihood(y, yp):
    return yp*y+(1-yp)*(1-y)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
#From 0 to 1  and 0.1 variation
Y = np.arange(0,1,0.01)
YP = np.arange(0,1,0.01)
Y, YP = np.meshgrid(Y, YP)
Z = likelihood(Y, YP)

surf = ax.plot_surface(Y, YP, Z, cmap=cm.coolwarm)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
#Real and prediction meet maximum at 0 and 1
```
